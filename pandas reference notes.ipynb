#### The Pandas library is built on NumPy and provides easy-to-use data structures and data analysis tools for the Python programming language. 

```python
import pandas as pd
```

Data Frame is a 2-D data structure with columns capable of holding different data types. 
Creating a Data Frame - 

```python
df = pd.DataFrame({'Yes': [50, 21], 'No': [131, 2]})
df
```

#### Series is a 1-D data structure capable of holding any data type. Creating a series - 

```python
pd.Series([3, -5, 7, 4], index=['a', 'b', 'c', 'd'])
```

#### Accessing and modifying a dataframe

```python
print('Reading column Yes, row 0 :', df['Yes'][0])
df['Yes'][1] = 25
print('Reading column Yes, row 1 :', df['Yes'][1])
```

#### Reading and writing from a csv

```python
data = pd.read_csv('data/uk-postcode.csv')
df.to_csv('data/myDataFrame.csv')
data
```

```python
data[3100:] # accessing data after row number 3100
```

```python
data[:-5] # ignoring the last 5 rows
```

#### The list of row labels used in a DataFrame is known as an Index. We can assign values to it by using an index parameter in our constructor:

```python
pd.DataFrame({'Bob': ['I liked it.', 'It was awful.'], 'Sue': ['Pretty good.', 'Bland.']}, index=['Product A', 'Product B'])
```

```python
series = pd.Series([30, 35, 40], index=['2015 Sales', '2016 Sales', '2017 Sales'], name='Product A')
series['2015 Sales']
```
